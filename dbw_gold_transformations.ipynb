### Import the pyspark lib's

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### To Read the Data

```python

df_gld = spark.read.format("parquet")\
    .option("header", "true")\
        .option("inferSchema", "true")\
        .load("abfss://silver-prime-clean@apsadatastore.dfs.core.windows.net/amazon_prime_titles_clean.csv")
df_gld.display()
```

### To view the schema

```python
df_gld.printSchema()
```

### Convert date into proper date formart mm/dd/yyyy
### Convert date into proper date added and year added formart

```python
df_gld = df_gld.withColumn("date_added", to_date(df_gld["date_added"], "MM/dd/yyyy"))
df_gld = df_gld.withColumn("year_added", year(df_gld["date_added"]))
df_gld.display()
```

```python
df_gld = df_gld.withColumn("category_1", split(df_gld["listed_in"], ",")[0])
df_gld = df_gld.withColumn("category_2", split(df_gld["listed_in"], ",")[1])
df_gld.display()
```

Replacing null values with Unkown

```python
df_gld = df_gld.withColumn("country_2",when(df_gld["category_2"].isNull(), "Unknown").otherwise(df_gld["category_2"]))
df_gld.display()
```

saving final report to 

```python
df_gld.write.format("delta")\
    .mode("append")\
        .option("path", "abfss://gold-prime-report@apsadatastore.dfs.core.windows.net/amazon_prime_titles_report.csv")\
            .save()
```
